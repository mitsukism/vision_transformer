```python
import shutil
root='/content/drive/MyDrive/ch3/data/FractalDB-1k/fractaldb_cat60_ins1000'
# フォルダをZIP形式で圧縮（'/content/drive/MyDrive/your_folder'は圧縮したいフォルダのパス）
for i in range(len(items)):
  shutil.make_archive(root+items[i]+'_compressed', 'zip', root+items[i])
```

```python
extract_zip = [
    'crystal',          'koch_curve',        'a2',
    'a3',               'maple_leaf',        'a4',
    'drumlin',           'mcWorter_pedigree', 'spiral',
    'spleenwort_fern',   'bamboofern',        'newyorkfern',
    'bedder',            'octopuslegs',       'flower',
    'branch',            'gaku',              'supernova',
    'rose',              'tree',              'grassfern',
    'sierpinski_carpet', 'greygoldenrod',     'groundpine',
    'sierpinski_pentagon','watersprite',      'coral',
    'involucre',         'snail',             'zigzag'
]
```

```python
import zipfile
for i in range(len(extract_zip)):
  # ZIPファイルのパスと解凍先のパスを指定
  zip_path = '/content/drive/MyDrive/ch3/data/FractalDB-1k/fractaldb_cat60_ins1000/'+'fractaldb_cat60_ins1000'+ extract_zip[i]+'_compressed.zip'
  extract_path = '/content/drive/MyDrive/ch3/data/FractalDB-1k/fractaldb_cat60_ins1000/'+extract_zip[i]

  with zipfile.ZipFile(zip_path, 'r') as zip_ref:
      zip_ref.extractall(extract_path)
```
